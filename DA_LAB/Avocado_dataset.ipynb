```python
import pandas as pd

data = pd.read_csv('avocado_10thQuestion.csv')
display(data)
```

# a) How to identify the unique values in the region column.

```python
Region_unique = data['region'].unique()
pf = pd.DataFrame(Region_unique)
display(pf)
```

# b) What is the maximum price for an avocado in the dataset.


```python

max = data['AveragePrice'].max()
print("the maximum price of the avocado:",max)
```

# c) Identify the type distribution and take a single avocado in the dataset and find out the median price ,mean,and standard deviation.

```python
type_distribution = data['type'].value_counts()
print("Type Distribution:")
print(type_distribution)
single_avocado = data.iloc[0]
median_price = single_avocado['AveragePrice']
mean_price = single_avocado['AveragePrice']
std_dev_price = 0  

print("\nStatistics for a Single Avocado:")
print("Median Price:", median_price)
print("Mean Price:", mean_price)
print("Standard Deviation:", std_dev_price)
```

# d)Find the highest,lowest price for conventional avocado’s in year with location.


```python

conven_avocados = data[data['type'] == 'conventional']
highest_prices = conven_avocados.groupby(['year', 'region'])['AveragePrice'].max()
lowest_prices = conven_avocados.groupby(['year', 'region'])['AveragePrice'].min()

print("Highest Prices for Conventional Avocados:")
display(highest_prices)
print("Lowest Prices for Conventional Avocados:")
display(lowest_prices)
```

# e)Draw the plots of the distribution of average price for different types of Avocados

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='type', y='AveragePrice', data=data)
plt.title('Distribution of Average Prices for Different Types of Avocados')
plt.xlabel('Avocado Type')
plt.ylabel('Average Price')
plt.show()
```

# f) Find the correlation matrix to measure the strength of the correlation between variables

```python

columns_correlation = ['AveragePrice', 'Total Volume', '4046', '4225', '4770']
correlation_data = data[columns_correlation]
correlation_matrix = correlation_data.corr()
print("Correlation Matrix:")
display(correlation_matrix)
```

# g) Findout the volume of avocado sales has increased in the last 5 years.

```python
data['Date'] = pd.to_datetime(data['Date'])

data['Year'] = data['Date'].dt.year

total_volume_by_year = data.groupby('Year')['Total Volume'].sum()
LA_5_YR = total_volume_by_year.index[-5:]

print("Total Volume of Avocado Sales for the Last 5 Years:")
display(total_volume_by_year[LA_5_YR])
```
