# National Universities Rankings.

```python
import pandas as pd

data = pd.read_csv("National_Universities_Rankings.csv")

display(data)
```

### a. Find the universities along with state whose fee is in between 25,000 to 30,000.

```python
data["TuitionFees"] = [int("".join(filter(str.isdigit, s))) for s in data["Tuition and fees"]]

df = data[(data["TuitionFees"] >= 25000) & (data["TuitionFees"] <= 30000)]

display(df[['Name', 'Location', 'Rank', 'Tuition and fees']])
```

### b. Find university where undergraduate enrollement is more than 25000 containing in-state students.

```python
data["UndergradEnrollment"] = [int("".join(filter(str.isdigit, s))) for s in data["Undergrad Enrollment"]]

df = data[data['UndergradEnrollment'] > 25000]

display(df[['Name', 'Location', 'Rank', 'In-state', 'Undergrad Enrollment']])
```

### c. Find the states where universities are located in three or more cities.

```python
data["state"] = [i[-2:] for i in data['Location']]

city_counts = data.groupby("state")["Name"].count()

display(city_counts[city_counts >= 3].reset_index())
```

### d. Find max and min tuition fee in each state.

```python
display(data.groupby('state').agg({'Tuition and fees': ['max', 'min']}))
```

### e. Find the city and state where maximum tuition fee difference in the universities in that city is greater than 5,000.

```python
df = data.groupby('Location').agg({'TuitionFees': ['max', 'min']})

display(df[(df[("TuitionFees","max")] - df[("TuitionFees","min")]) > 5000])
```

### f. Identify the correlation between Enrollment Number and University Rank.

```python
print("The correlation between Enrollment Number and University Rank is:",data["UndergradEnrollment"].corr(data["Rank"]))
```

### g. Print university name and where it is located.

```python
display(data[["Name","Location"]])
```

### h. Find cities having more than two universities along with state.

```python
city_counts = data.groupby("Location")["Name"].count()

display(city_counts[city_counts > 2].reset_index())
```

### i. Find the number of states and cities locating top 100 universities.

```python
print(f"Number of states locating top 100 universities: {data.head(100)['state'].nunique()}")
print(f"Number of cities locating top 100 universities: {data.head(100)['Location'].nunique()}")
```

### j. Draw the plot to show undergraduate enrollment of each university.

```python
import matplotlib.pyplot as plt

plt.bar(data.head(5)['Name'], data.head(5)['UndergradEnrollment'])

plt.xlabel('University')
plt.ylabel('Undergraduate Enrollment')
plt.title('Undergraduate Enrollment of Each University')

plt.xticks(rotation=30, ha="center")
plt.show()
```

### k. Draw the plot to show university name and its corresponding tuition fee.

```python
import matplotlib.pyplot as plt

plt.bar(data.head(5)['Name'], data.head(5)['TuitionFees'])

plt.xlabel('University')
plt.ylabel('Tuition Fees')
plt.title('Tuition Fees of Each University')

plt.xticks(rotation=30, ha="center")
plt.show()
```

### l. Plot the number of universities in each state having ranks > 100.

```python
import matplotlib.pyplot as plt

filtered_data = data[data['Rank'] > 100]

state_counts = filtered_data.groupby('state').size()

plt.figure(figsize=(15, 6))
state_counts.plot(kind='bar')

plt.xlabel('State')
plt.ylabel('Number of Universities (Rank > 100)')
plt.title('Number of Universities in Each State')

plt.xticks(rotation=0, ha='center')
plt.show()
```
